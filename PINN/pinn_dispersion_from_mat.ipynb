```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

out = Path('../Result/dispersion_plots')
files = [
    out / '01_time_space_field.png',
    out / '02_time_traces.png',
    out / '03_fk_map.png',
    out / '04_dispersion_ridge.png',
]

fig, axes = plt.subplots(2,2, figsize=(12,8))
for ax, fp in zip(axes.ravel(), files):
    if fp.exists():
        ax.imshow(mpimg.imread(fp))
        ax.set_title(fp.name)
    else:
        ax.text(0.5,0.5,f'Missing: {fp.name}',ha='center',va='center')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
