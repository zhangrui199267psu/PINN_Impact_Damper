# Parametric PINN (Force-only): sweep $\phi_1$ and $\phi_2$

This notebook runs parametric PINN response simulations where only force parameters vary:
- `phi1` = forcing amplitude
- `phi2` = forcing frequency

All structural and solver settings are kept constant.

```python
from pathlib import Path
import sys

ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'parametric_PINN'))

from parametric_pinn_force_sweep import (
    StructuralConfig,
    TrainingConfig,
    run_parametric_sweep,
)
```

```python
# Keep all structural settings fixed
structural_cfg = StructuralConfig(
    n_dof=20,
    m_x=1.0,
    m_y=0.3,
    k=1.0,
    c=0.0,
    D=1.0,
    r=1.0,
)

training_cfg = TrainingConfig(
    num_neurons=64,
    beta_icx=1.0,
    beta_fx=1.0,
    optimizer_lb=True,
    seg_window=1.0,
    n_points_per_seg=1000,
    n_iter_per_seg=1000,
    n_segments=50,
)
```

```python
# Vary ONLY phi1 and phi2
phi_cases = [
    (0.1, 0.5),
    (0.1, 1.0),
    (0.1, 2.0),
    (0.3, 1.0),
]

run_parametric_sweep(
    phi_cases=phi_cases,
    s_cfg=structural_cfg,
    t_cfg=training_cfg,
    out_dir=ROOT / 'Result',
)

print('Done. Check Result/ for pinn_data_phi1_*_phi2_*.mat files.')
```
